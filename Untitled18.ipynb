```python
import pandas as pd
import numpy as np
df = pd.read_csv("IBM HR-dataset.csv")
```

```python
print("ðŸ”¹ First 5 Rows:")
print(df.head())
print("\nðŸ”¹ Shape of Dataset:", df.shape)
print("\nðŸ”¹ Data Types:")
print(df.dtypes)
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())
```

```python
df.dropna(axis=1, how='all', inplace=True)    #drop empty rows
```

```python
for col in df.columns:                        #fill missing numeric values with median & categiorical with mode
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

print("\nâœ… Missing values handled successfully!")
```

```python
duplicates = df.duplicated().sum()        #remove duplicates
print(f"\nðŸ”¹ Duplicates found: {duplicates}")
df.drop_duplicates(inplace=True)
```

```python
cols_to_drop = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber']
df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True, errors='ignore')
print("\nâœ… Irrelevant columns removed (if any).")
```

```python
print("\nðŸ“ˆ Final Dataset Shape:", df.shape)
print("\nðŸ“Š Final Data Sample:")
print(df.head())
```

```python
df.to_csv("IBM_HR_Cleaned.csv", index=False)
print("\nðŸ’¾ Cleaned dataset saved as 'IBM_HR_Cleaned.csv'")
```

```python
#EDA
#OUTLIERS
import matplotlib.pyplot as plt

def plot_outliers(df, column):
    plt.figure(figsize=(6,4))
    plt.boxplot(df[column].dropna())
    plt.title(f"Boxplot - Outliers in {column}")
    plt.xlabel(column)
    plt.show()
```

```python
plot_outliers(df, "TotalWorkingYears")
```

```python
plot_outliers(df, "MonthlyIncome")
```

```python
#REMOVE OUTLIERS
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clean = df[(df[column] >= lower) & (df[column] <= upper)]
    return df_clean
```

```python
df_no_outlier = remove_outliers(df, "TotalWorkingYears")

print("Before:", df.shape)
print("After :", df_no_outlier.shape)
```

```python
plot_outliers(df_no_outlier, "TotalWorkingYears")
```

```python
#univariate visuals
# 1 Attrition count
plt.figure(figsize=(6,4))
if 'Attrition_binary' in df.columns:
    df['Attrition_binary'].value_counts().sort_index().plot(kind='bar')
    plt.title('Attrition Count (0=Stay,1=Left)')
    plt.ylabel('Count')
else:
    df['Attrition'].value_counts().plot(kind='bar')
```

```python
# 2 Age distribution
plt.figure(figsize=(8,4))
df['Age'].dropna().plot(kind='hist', bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
```

```python
# 3 MonthlyIncome raw
plt.figure(figsize=(8,4))
df['MonthlyIncome'].dropna().plot(kind='hist', bins=40)
plt.title('MonthlyIncome (raw)')
plt.xlabel('MonthlyIncome')
```

```python
#bivariate visual box and bar plots
# 1 JobSatisfaction counts (if present)
if 'JobSatisfaction' in df.columns:
    plt.figure(figsize=(6,4))
    df['JobSatisfaction'].value_counts().sort_index().plot(kind='bar')
    plt.title('JobSatisfaction Counts')
```

```python
# 2 Attrition rate by Department
df["Attrition_binary"] = df["Attrition"].map({"Yes":1, "No":0})


dept_rate = df.groupby("Department")["Attrition_binary"].mean().sort_values(ascending=False)

plt.figure(figsize=(7,4))
dept_rate.plot(kind='bar')

plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Department")
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# 3 Attrition rate by JobRole
if 'JobRole' in df.columns and 'Attrition_binary' in df.columns:
    job_rate = df.groupby('JobRole')['Attrition_binary'].mean().sort_values(ascending=False)
    plt.figure(figsize=(10,5))
    job_rate.plot(kind='bar')
    plt.ylabel('Attrition Rate'); plt.title('Attrition Rate by JobRole')
    plt.xticks(rotation=45, ha='right')
```

```python
# 4 OverTime vs Attrition stacked counts
if 'OverTime' in df.columns and 'Attrition_binary' in df.columns:
    pivot = pd.crosstab(df['OverTime'], df['Attrition_binary'])
    plt.figure(figsize=(6,4))
    pivot.plot(kind='bar', stacked=True)
    plt.title('OverTime vs Attrition (counts)')
```

```python
# 5 MonthlyIncome by Attrition (boxplot)
if 'MonthlyIncome' in df.columns and 'Attrition_binary' in df.columns:
    plt.figure(figsize=(8,4))
    df.boxplot(column='MonthlyIncome', by='Attrition_binary')
    plt.title('MonthlyIncome by Attrition'); plt.suptitle('')
```

```python
#multivariate visual
# 1 YearsAtCompany by Attrition (boxplot)
if 'YearsAtCompany' in df.columns and 'Attrition_binary' in df.columns:
    plt.figure(figsize=(8,4))
    df.boxplot(column='YearsAtCompany', by='Attrition_binary')
    plt.title('YearsAtCompany by Attrition'); plt.suptitle('')
```

```python
# 2 Correlation heatmap (numeric)
import seaborn as sns
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=['int64','float64']).corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Matrix (numeric)')
```

```python
# 3 Scatter: MonthlyIncome vs TotalWorkingYears colored by Attrition
if set(['MonthlyIncome','TotalWorkingYears','Attrition_binary']).issubset(df.columns):
    plt.figure(figsize=(7,5))
    stayed = df[df['Attrition_binary']==0]
    left = df[df['Attrition_binary']==1]
    plt.scatter(stayed['TotalWorkingYears'], stayed['MonthlyIncome'], alpha=0.5, label='Stay')
    plt.scatter(left['TotalWorkingYears'], left['MonthlyIncome'], alpha=0.6, marker='x', label='Left')
    plt.xlabel('TotalWorkingYears'); plt.ylabel('MonthlyIncome'); plt.title('MonthlyIncome vs TotalWorkingYears')
    plt.legend()
```

```python
# 4 Pivot heatmap: Department vs JobRole attrition rate
if set(['Department','JobRole','Attrition_binary']).issubset(df.columns):
    pivot = df.pivot_table(index='Department', columns='JobRole', values='Attrition_binary', aggfunc='mean')
    plt.figure(figsize=(12,4))
    sns.heatmap(pivot.fillna(0), annot=True, fmt='.2f', cmap='YlOrRd')
    plt.title('Attrition Rate: Department vs JobRole')
```

```python
# 5 TrainingTimesLastYear distribution (univariate / outlier check)
if 'TrainingTimesLastYear' in df.columns:
    plt.figure(figsize=(8,4))
    df['TrainingTimesLastYear'].dropna().plot(kind='hist', bins=30)
    plt.title('TrainingTimesLastYear Distribution')
```
